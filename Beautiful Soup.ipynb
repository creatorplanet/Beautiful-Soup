```python
from urllib.parse import urljoin
# href 주소가 제각각이므로, urljoin을 사용하여 절대경로로 잡힌 url은 그대로 두고, 상대경로로 잡힌 url은 절대경로로 변경한다.
 
rank = []  
main_menu = [] # 가게 메인 메뉴
cafe_name = [] # 가게 이름
url_add = []

list_soup = soup.find_all('div', 'sammy')
url_base = 'http://www.chicagomag.com'
for item in list_soup:
    rank.append(item.find(class_='sammyRank').get_text())
    
    tmp_string = item.find(class_='sammyListing').get_text()

    main_menu.append(re.split(('\n|\r\n'), tmp_string)[0])
    cafe_name.append(re.split(('\n|\r\n'), tmp_string)[1])
    
    url_add.append(urljoin(url_base, item.find('a')['href']))
```

```python
from tqdm.notebook import tqdm
from urllib.request import Request, urlopen

price = []
address = []

for n in tqdm(df.index):
    req = Request(df['URL'][n], headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(req)
    soup_tmp = BeautifulSoup(html, 'lxml')
    
    gettings = soup_tmp.find('p', 'addy').get_text()
    
    price.append(gettings.split()[0][:-1])
    address.append(' '.join(gettings.split()[1:-2]))
```

```python
import folium
import pandas as pd
import googlemaps
import numpy as np
```

```python
df = pd.read_csv('./data/03. best_sandwiches_list_chicago2.csv', index_col=0)
df.head(5)
```

```python
gmaps_key = "***************************" # Google Maps Geocoding API key 
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
# 50개 맛집의 위도, 경도 정보를 받아온다.

lat = []
lng = []
for n in tqdm(df.index):
    if df['Address'][n] != 'Multiple':
        target_name = df['Address'][n]+', '+'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
        
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
len(lat), len(lng)
```

```python
df['lat'] = lat
df['lng'] = lng
df.head()
```

```python
# 주소에 Multiple이 나타나지 않는 경우만 주소를 검색하도록 한다.
# 50개 맛집의 위도, 경도의 평균값을 중앙에 둔다. 
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], 
                                      zoom_start=11)
folium.Marker([df['lat'].mean(), df['lng'].mean()], 
                                      popup='center').add_to(mapping)
mapping
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], 
                     zoom_start=11)

for n in df.index:
    if df['Address'][n] != 'Multiple':
        folium.Marker([df['lat'][n], df['lng'][n]], 
                                      popup=df['Cafe'][n]).add_to(mapping)

mapping
```

# 네이버 영화 평점 기준 영화의 평점 변화 확인하기

```python
from bs4 import BeautifulSoup  
import pandas as pd
from urllib.request import Request, urlopen
req = Request("https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20170806", headers={'User-Agent': 'Mozilla/5.0'})
page = urlopen(req)
soup = BeautifulSoup(page, "html.parser")
soup
```

```python
movie_name = [soup.find_all('div', 'tit5')[n].a.string for n in range(0, 47)]
movie_name
```

```python
movie_point = [soup.find_all('td', 'point')[n].string for n in range(0, 47)]
movie_point
```

```python
# 5월 1일부터 100일간으로 날짜를 정의한다.
date = pd.date_range('2017-5-1', periods=100, freq='D')
date
```

```python
len(date)
```

```python
import urllib
from tqdm.notebook import tqdm

movie_date = []
movie_name = []
movie_point = []

for today in tqdm(date): 
    req = Request('https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date={date}'.format(date=urllib.parse.quote(today.strftime('%Y%m%d'))), headers={'User-Agent': 'Mozilla/5.0'})
    response = urlopen(req)
    soup = BeautifulSoup(response, "html.parser")
    
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
```

```python
len(movie_date), len(movie_name), len(movie_point)
```

```python
movie = pd.DataFrame({'date':movie_date, 'name':movie_name, 'point':movie_point})
movie.head()
```

```python
movie.info()
```

```python
movie['point'] = movie['point'].astype(float)
movie.info()
```

```python
# 영화별로 점수의 합산을 데이터로 보고 싶을 경우, pivot_table을 사용한다.
# aggfunc으로 np.sum을 이용해서 합산을 해야 영화별 점수의 합계로 정렬될 것이다.
# 고득점 영화 1위~5위
import numpy as np

movie_unique = pd.pivot_table(movie, index=['name'], aggfunc=np.sum)
movie_best = movie_unique.sort_values(by='point', ascending=False)
movie_best.head()
```

```python
# 날짜별 평점의 변화를 확인할 수 있다.
tmp = movie.query('name == ["라라랜드"]')
tmp
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))
plt.plot(tmp['date'], tmp['point'])
plt.legend(loc='best')
plt.grid()
plt.show()
```
